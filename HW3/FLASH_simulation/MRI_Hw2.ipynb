```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

n = np.arange(0.1, 1.1, 0.1)
z = np.arange(0, 1, 0.1)
from math import *
x = 1 #Mz(0)
m = 1 #M0
angle = 25. #flip angle
t = 0.1 #TR/T1
y1 = x*np.cos(angle/180*pi)
y2 = y1*np.exp(-t*n)+m*(1-np.exp(-t*n))

for ii in range(99):
    x = (x*np.cos(angle/180*pi)*np.exp(-t))+m*(1-np.exp(-t))
y3 = x*np.cos(angle/180*pi)
y4 = y3*np.exp(-t*n)+m*(1-np.exp(-t*n))

plt.subplot(121)
plt.plot(z, y2)
plt.grid()
plt.subplot(122)
plt.plot(z, y4)
plt.grid()

y2 = y2[-1]
y4 = y4[-1]

print ("Mz': %f , Mz: %f" % (y1, y2))
print ("Mz': %f , Mz: %f" % (y3, y4))
```
